```python
# вспоминаем как работать со списками и словарями
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
for key in geo_data:
    print (key, geo_data[key])
    for key2 in geo_data[key]:
        print (key2)
```

```python
def geografier(keyword):
    """сплитуем киворд и ищем геодату"""
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    region = None #не очень понятн оверно ли я присвоил тип Null 
    key_string = keyword.split(' ')
    for key_from_geo in geo_data:
        for key_keyword in key_string:
            if key_keyword in geo_data[key_from_geo]:
                region = key_from_geo
    return region
        

import pandas as pd
data = pd.read_csv('keywords.csv')
data['region'] = data['keyword'].apply(geografier)
data = data.sort_values(by='region', ascending=True)
data
```
